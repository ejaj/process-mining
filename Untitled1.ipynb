```python
loaded_df = pd.read_csv("students_data.csv")
loaded_df.head(5)
```

```python
# Transform the dataset into an event log format
event_log = loaded_df.rename(columns={
    'STUDIENR': 'Case ID',
    'KURSTXT': 'Activity',
    'BEDOMMELSESDATO': 'Timestamp',
    'BEDOMMELSE': 'Grade',
    'ECTS': 'ECTS'
})

 

# Convert Timestamp to datetime format and sort
event_log['Timestamp'] = pd.to_datetime(event_log['Timestamp'])
event_log = event_log.sort_values(by=['Case ID', 'Timestamp'])
# Display the first few rows of the transformed event log
event_log.head()
```

```python
df_sorted = event_log.sort_values(by='Timestamp')
df_sorted.head()
```

```python
event_log.to_csv('students_data_n new_date.csv', index=False)
```
